```python
from newspaper import Article, ArticleException
def extract_text(url, index):
  try:
      article=Article(url)
      article.download()
      article.parse()
      return article.text
  except ArticleException as e:
      print(f"Failed to extract text from URL at index{index}: {url}. Error: {e}")
      return None
```
