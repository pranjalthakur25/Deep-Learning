```python
import numpy
```

```python
#sop is sum of product
#sop=x1w
def pranjal_sigmoid(sop):
  return 1.0/(1+numpy.exp(-1*sop))
```

```python
def error(predicted, target):
  return numpy.power(predicted-target, 2)
```

```python
x=0.5
target=0.8
learning_rate=0.001
w=numpy.random.rand()
```

```python

#Forward Pass
y=w*x
predicted=pranjal_sigmoid(y)
err=error(predicted, target)
```

```python
print(predicted)
```

```python
#x^2=2*x i.e. derivative of error
def derivative(predicted, target):
  return 2*(predicted- target)
```

```python
#derivative of sigmoid activation function
def activation__sop_derivation(sop):
  return pranjal_sigmoid(sop)*(1.0-pranjal_sigmoid(sop))
```

```python
def stop_w_derive(x):
  return x
```

```python
def update_w(w, grad, learning_rate):
  return w-learning_rate*grad
```

```python
#backward pass
g1=error(predicted, target)
g2=activation__sop_derivation(predicted)
g3=stop_w_derive(x)
grad=g3*g2*g1
```

```python
print(predicted)
```

```python
w=update_w(w, grad, learning_rate)
```
