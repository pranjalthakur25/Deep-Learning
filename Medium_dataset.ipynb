```python
df['title']=df['title'].apply(lambda x:x.replace('\xa0',' '))
df['title']=df['title'].apply(lambda x: x.replace('\u200a',' '))
```

```python
df['title'][1]
```

```python
text=df['title'].str.lower().str.split().explode()
```

```python
text
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer=Tokenizer()
```

```python
tokenizer.fit_on_texts(text.str.lower())
```

```python
tokenizer.word_index
```

```python
input_sequences=[]
for sentence in text.tolist():
  tokenized_sentence=tokenizer.texts_to_sequences([sentence])[0]
  # print(tokenized_sentence)
  for i in range(1,len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
max_len=max([len(x) for x in input_sequences])
```

```python
max_len
```

```python
from keras.utils import pad_sequences
padded_input_sequence=pad_sequences(input_sequences,maxlen=max_len,padding='pre')
```

```python
x=padded_input_sequence[:,:-1]
```

```python
x
```

```python
y=padded_input_sequence[:,-1]
```

```python
y.max()
```

```python
x.shape
```

```python
y.shape
```

```python
from tensorflow.keras.utils import to_categorical
y=to_categorical(y, num_classes=8227)
```

```python
y.shape
```

```python
y
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, GRU
```

```python
model=Sequential()
model.add(Embedding(8227,300,input_length=19))
model.add(LSTM(200))
model.add(Dense(8227, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history=model.fit(x,y,epochs=40)
```

```python
model_G=Sequential()
model_G.add(Embedding(8227,300,input_length=19))
model_G.add(GRU(200))
model_G.add(Dense(8227,activation='softmax'))
```

```python
model_G.summary()
```

```python
model_G.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history1=model_G.fit(x,y,epochs=40)
```

```python
model1=Sequential([
    Embedding(8227, 300, input_length=19),
    LSTM(150,return_sequences=True),
    LSTM(150),
    Dense(8227,activation='softmax')
])
model1.summary()
```

```python
model1.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
history3=model1.fit(x,y,epochs=10, batch_size=32,validation_split=0.2)
```

```python
model2=Sequential([
    Embedding(8227, 300, input_length=19),
    GRU(150,return_sequences=True),
    GRU(150),
    Dense(8227,activation='softmax')
])
model2.summary()
```

```python
model2.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```
