```python
df['title']=df['title'].apply(lambda x:x.replace('\xa0',' '))
df['title']=df['title'].apply(lambda x: x.replace('\u200a',' '))
```

```python
text=df['title'].str.lower().str.split().explode()
```

```python
text
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer=Tokenizer()
```

```python
tokenizer.fit_on_texts(text.str.lower())
```

```python
tokenizer.word_index
```

```python
input_sequences=[]
for sentence in text.str.split('\n'):
  tokenized_sentence=tokenizer.texts_to_sequences([sentence])[0]
  # print(tokenized_sentence)
  for i in range(1,len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
```
